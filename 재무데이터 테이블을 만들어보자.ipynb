```python
def stock_df(df):


    df = df.iloc[:,:13]
    df = df.sort_values(by='sj_div')
    df = df[df.fs_nm == '연결재무제표']
    df.reset_index(drop=True, inplace=True)

    df['누적_x'] = df['thstrm_amount'].apply(lambda x:x.replace(',',''))
    df['누적_x'] = df['누적_x'].astype('int64')


    return df



api_key = 'f09cfd7f19a3c92c3f8ddeba0993187c4ca93674'
# stock_names = ['005930'] # 삼성전자 종목코드
dart = OpenDartReader(api_key)

year_li = [2016, 2017, 2018, 2019, 2020, 2021]
year_df = pd.DataFrame()

for year in year_li[:]:

    df1 = dart.finstate(corp='005930', bsns_year=year, reprt_code=11013)
    df2 = dart.finstate(corp='005930', bsns_year=year, reprt_code=11012)
    df3 = dart.finstate(corp='005930', bsns_year=year, reprt_code=11014)
    df4 = dart.finstate(corp='005930', bsns_year=year, reprt_code=11011)


    df1 = stock_df(df1)
    df2 = stock_df(df2)
    df3 = stock_df(df3)
    df4 = stock_df(df4)


    for cost in range(8,12):
        df4.iloc[cost, -1] = df4.iloc[cost, -1] - df3.iloc[cost, -1]

    for cost in range(8,12):
        df3.iloc[cost, -1] = df3.iloc[cost, -1] - df2.iloc[cost, -1]

    for cost in range(8,12):
        df2.iloc[cost, -1] = df2.iloc[cost, -1] - df1.iloc[cost, -1]
    
    df = pd.concat([df1, df2, df3, df4])
    
    year_df = pd.concat([year_df, df])
    
df_2022 = dart.finstate(corp='005930', bsns_year=2022, reprt_code=11013)

df_2022 = stock_df(df_2022)

total_df = pd.concat([year_df, df_2022])

total_df
```
