# 공시 전처리

```python
df1 = pd.read_csv('./data/공시_다트크롤링/AJ네트웍스.csv',index_col=0)
df1.head()
```

```python
df1 = df1[['접수일자','공시대상회사','보고서명']]
df1.columns = ['날짜','공시회사','공시보고서명']
df1
```

```python
df1['날짜'] = pd.to_datetime(df1['날짜'])
df1['날짜'] = df1['날짜'].apply(lambda x : x.to_period(freq='D'))
```

```python
df1.drop_duplicates(inplace=True)
df1 = df1.sort_values(by='날짜')
df1
```

```python
df_a = df1.groupby('날짜')[['공시보고서명','공시회사']].sum().reset_index()
```
